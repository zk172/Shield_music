### Initial dump

```python
s3 = s1['Music;'].str.contains(r'Season\s\d\s',regex=True)
episode_names=s1['Music;'][s3]
episode_names

aaa= [[] for _ in range(episode_names.shape[0])]
aaa

for i in range(0,episode_names.shape[0]-1):
    aaa[i] = s1.loc[episode_names.index[i]:episode_names.index[i+1]-1,'Music;']

aaa[i+1] = s1.loc[episode_names.index[i+1]:,'Music;']
for a in aaa:
    print(a)
```

```python
len(aaa)
```

```python
num_songs = s1.shape[0]/3 -5
num_songs = np.round(num_songs,1).astype(int)
```

```python
music = [[] for _ in range(num_songs)]
j=0
i=3
k=0
#for j in range(0,10):
   # for i in aaa:
   #     music[j] = aaa[j][i:i+3]
     #  i=i+3

for j in range(0,len(aaa)):
    i=3
    if isinstance(aaa[j],pd.Series) == True:
        for i in range(3,aaa[j].shape[0],3):
            music[k] = aaa[j].iloc[i:i+3]
            k=k+1
    else:
        break
```

```python
df22 = pd.DataFrame(['','','']).T
for i in range(0,60):
    if isinstance(music[i],pd.Series) == True:
        song_name = music[i].iloc[0]
        artist = music[i].iloc[1]
        how_used = music[i].iloc[2]
        df24 = pd.DataFrame([song_name,artist,how_used]).T
        df22 = pd.concat([df22, df24],axis = 0, ignore_index=True)
    else:
        break
```
